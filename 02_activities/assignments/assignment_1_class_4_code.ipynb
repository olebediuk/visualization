```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)

#plot both variable on the same axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
fig.show()
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.random.randint([110,180,240,99,200])
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2, 
                              nrows=1, 
                              figsize=(7, 3))

font1 = {'family':'serif','size' :10}
font2 = {'family':'serif','size' : 10}

ax1.scatter(x1,y1,
            marker = 'o',
            color='red')
ax2.bar(x2,y2)

ax1.annotate('This is important!', xy=(5, 35), xytext=(10, 10), 
            arrowprops=dict(arrowstyle = "wedge",
                            color = "hotpink"))
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7, 4),
                                    layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
