```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set a random seed for reproducibility
np.random.seed(613)
# Generate x values from 0 to 49
x = np.arange(50)
# Generate 50 random integers between 0 and 100
y = np.random.randint(0, 100,50)
# Create a figure and a single subplot (ax), with a custom figure size
fig, ax = plt.subplots(figsize=(5, 3))
#ax.scatter(x,y)
#ax.bar(x,y)
#ax.barh(x,y)
#ax.plot(x,y)
# Histogram: shows the distribution of values in y
ax.hist(y)
```

```python
#Create custome font
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}



fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x, 
           y, 
           marker='+',  
           color = "indigo",) 
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x, y, marker='*',  
           color = '#7425b9',
           linestyle = '--' ,
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
ax.grid(axis = 'y',color = 'green',linewidth = 4, linestyle = '--')
```
